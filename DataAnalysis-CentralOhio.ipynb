```python
#Operating Expenditures vs Enrollment data for 2016-2017

import pandas as pd

centralOhioSchools = ["Bexley City", "Big Walnut Local", "Buckeye Valley Local", "Canal Winchester Local", 
                      "Columbus City School District", "Delaware City", "Dublin City", "Fairbanks Local",
                      "Gahanna-Jefforson City", "Grandview Heights Schools City", "Groveport Madison Local", 
                      "Hamilton City", "Hilliard City", "Jefferson Local", "Johnstown-Monroe Local",
                      "Jonathan Alder Local", "Licking Heights Local", "Marysville Exempted Village", 
                      "New Albany-Plain Local", "Northridge Local", "Olentangy Local", "Pickerington Local",
                      "Reynoldsburg City", "South-Western City", "Southwest Licking Local", "Teays Valley Local",
                      "Upper Arlington City", "Westerville City", "Westfall Local", "Whitehall City", "Worthington City"]

districts = pd.read_excel('Data/Raw Data/FY17_EFM_Data.xls').dropna(how = 'all')
enrollment = pd.read_excel('Data/Raw Data/1617_DIST_ENROLL.xls').dropna(how = 'all')

#Rename some stuff
districts.rename(columns = {'Org Type' : 'OrgType', 'Local Education Agency Name' : 'District', 
                     'Operating Expenditures' : 'OperatingExpenditures'}, inplace = True)
enrollment.rename(columns = {'District Name' : 'District'}, inplace = True)

#Drop everything that isn't a public school district
districts = districts[districts.OrgType == "Public District"]

#Get only central Ohio schools
districts = districts[districts['District'].isin(centralOhioSchools)]

districts = districts.join(enrollment.set_index('District'), on='District')

#Drop unnecessary data
districts = districts.drop(['IRN', 'Instruction', 'Pupil Support', 'Staff Support', 'Operations Support', 'Operating EPP',
                  'Instruction EPP', 'Pupil Support EPP', 'Staff Support EPP', 'Administration EPP', 'Operations Support EPP',
                 'Operating EPEP', 'Instruction EPEP', 'Pupil Support EPEP', 'Staff Support EPEP', 'Administration',
                 'Administration EPEP', 'Operations Support EPEP', 'OrgType', 'District IRN', 'Phone #'], axis=1)

districts.to_csv('Data/CentralOhio/FundingVsEnrollment.csv', index = False)
```
